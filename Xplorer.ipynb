```python
import statsmodels.api as sm

# Define the columns for cross-plots
columns = ['MD', 'GR', 'RES.1', 'DT', 'RHOB', 'NEU', 'ΔLogR', 'Measured-TOC']

# Create separate scatter plots with linear regression lines and equations
for col1 in columns:
    for col2 in columns:
        if col1 != col2:  # Exclude plotting a column against itself
            # Fit the linear regression model
            X = df[col1]
            y = df[col2]
            X = sm.add_constant(X)  # Add constant term for intercept
            model = sm.OLS(y, X).fit()

            # Plot the data and regression line
            sns.lmplot(x=col1, y=col2, data=df, line_kws={'color': 'red'}, ci=None)

            # Plot the regression equation
            plt.text(df[col1].min(), df[col2].max(), f'y = {model.params[0]:.2f} + {model.params[1]:.2f}x', fontsize=12)

            # Set plot title and labels
            plt.title(f'{col1} vs {col2}')
            plt.xlabel(col1)
            plt.ylabel(col2)
            plt.show()
```
